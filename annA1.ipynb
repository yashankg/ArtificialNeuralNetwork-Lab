```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
#Sigmoid
x=np.linspace(-10, 10, 1000)
y=1/(1+np.exp(-x))
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Sigmoid")
plt.grid(True)
plt.plot(x,y)
plt.show()
```

```python
#tanh
x=np.linspace(-4, 4, 500)
y=2/(1+np.exp(-2*x))-1
plt.title("Tanh")
plt.grid(True)
plt.plot(x, y)
```

```python
#relu
x=np.linspace(-4, 4, 500)
y=np.maximum(0, x)
plt.title("Relu")
plt.grid(True)
plt.plot(x, y)
plt.show()
```

```python
#elu
x=np.linspace(-10, 10, 1000)
y=np.where(x>0, x, 1*(np.exp(x)-1))
plt.title("Elu")
plt.grid(True)
plt.plot(x, y)
plt.show()
```

```python
#binary step function
x=np.linspace(-10, 10, 1000)
y=np.where(x>0, 1, 0)
plt.title("Binary Step Function")
plt.grid(True)
plt.plot(x, y)
plt.show()
```
